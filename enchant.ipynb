```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
   UNBOLD = '\033[0m'


print(color.UNBOLD + 'Hello World !' + color.END)
```

```python
def bold_half(word):
  first_half = word[:len(word)//2]
  second_half = word[len(word)//2:]
  first_half_b = color.BOLD + first_half
  second_half_b = color.UNBOLD + second_half
  #print(first_half_b,second_half)
  return first_half_b + second_half_b
```
